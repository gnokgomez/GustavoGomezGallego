```python
x_train = np.array(Rlineal['metro']).reshape((-1,1))
y_train = np.array(Rlineal['precio'])

model = LinearRegression()
model.fit(x_train,y_train)

w=model.coef_
b=model.intercept_

print(f"interceptor b: {model.intercept_}")
print(f"pendiente w: {model.coef_}")
```

```python
x=np.linspace(0, Rlineal['metro'].max()+20,100)
y= w*x+b

Rlineal.plot.scatter(x='metro',y='precio')
plt.plot(x,y, '-g')
plt.ylim(0,Rlineal['precio'].max()*1.1)
plt.show()
```
